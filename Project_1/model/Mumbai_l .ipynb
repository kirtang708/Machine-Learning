```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('Location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft <= (m+st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
        
    return df_out



df5 = remove_pps_outliers(df4)
df5.shape
```

```python
def plot_scatter_chart(df, Location):
    bed2 = df[(df.Location == Location) & (df.No_of_Bedrooms == 2)]
    bed3 = df[(df.Location == Location) & (df.No_of_Bedrooms == 3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bed2.Area, bed2.Price, color='blue', label='2 BedRoom', s=50)
    plt.scatter(bed3.Area, bed3.Price, marker='+',color='red', label='3 BedRoom', s=50)
    plt.xlabel("Total sqare feet area")
    plt.xlabel("Price per square feet")
    plt.title(Location)
    plt.legend()

plot_scatter_chart(df5, 'Kharghar')
```

```python
def remove_bed_outliers(df):
    exclude_indices = np.array([])
    for Location, Location_df in df.groupby('Location'):
        No_of_Bedrooms_statistics = {}
        for No_of_Bedrooms, No_of_Bedrooms_df in Location_df.groupby('No_of_Bedrooms'):
            No_of_Bedrooms_statistics[No_of_Bedrooms] = {
                'mean' : np.mean(No_of_Bedrooms_df.price_per_sqft),
                'std' : np.std(No_of_Bedrooms_df.price_per_sqft),
                'count' : No_of_Bedrooms_df.shape[0]
            }
            
        for No_of_Bedrooms, No_of_Bedrooms_df in Location_df.groupby('No_of_Bedrooms'):
            statistics = No_of_Bedrooms_statistics.get(No_of_Bedrooms-1)
            if statistics and statistics['count']>5:
                exclude_indices = np.append(exclude_indices, No_of_Bedrooms_df[No_of_Bedrooms_df.price_per_sqft<(statistics['mean'])].index.values)
                
            
    return df.drop(exclude_indices, axis='index')

df6 = remove_bed_outliers(df5)
df6.shape
```

```python
df7 = df6.drop(['price_per_sqft'],axis='columns')
df7.head()
```

```python
dummies = pd.get_dummies(df7.Location)
dummies.head()
```

```python
df8 = pd.concat([df7,dummies.drop('other',axis='columns')],axis='columns')
df8.head()
```

```python
df9 = df8.drop('Location', axis='columns')
df9.head()
```

```python
df9.shape
```

```python
X = df9.drop('Price', axis='columns')
X.head()
```

```python
y = df9.Price
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train, y_train)
lr_clf.score(X_test, y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X,y,cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor


def find_best_model_using_gridsearchcv(X,y):
    algos={
        'linear_regression' : {
            'model' : LinearRegression(),
            'params':{
                'normalize':[True, False]
            }
        },
        'lasso': {
            'model':Lasso(),
            'params':{
                'alpha':[1,2],
                'selection':['random', 'cyclic']
            }
        },
        'decision_tree':{
            'model': DecisionTreeRegressor(),
            'params':{
                'criterion' : ['mse', 'friedman_mse'],
                'splitter':['best', 'random']
            }
        }
    }
    
    scores = []
    
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model':algo_name,
            'best_score':gs.best_score_,
            'best_params':gs.best_params_
        })
        
    return pd.DataFrame(scores, columns = ['model', 'best_score', 'best_params'])
    
find_best_model_using_gridsearchcv(X,y)
```

```python
def predict_price(Location, Area, No_of_Bedrooms, No_of_BathRooms, Gymnasium, Security ):
    loc_index = np.where(X.columns==Location)[0][0]
    
    x = np.zeros(len(X.columns))
    x[0] = Area
    x[1] = No_of_Bedrooms
    x[2] = No_of_BathRooms
    x[3] = Gymnasium
    x[4] = Security
    
    
    if loc_index >= 0:
        x[loc_index] = 1
        
    return lr_clf.predict([x])[0]
```
